```python
def size_changes(x):
    if 'M' in x:
        val=float(x.replace('M',''))
        val=val*1000
    elif 'k' in x:
        val = float(x.replace('k',''))
    else:
        val=0
    return val    
```

```python
df_dropna['Size']=df_dropna['Size'].apply(size_changes)
```

```python
def abc(x):
    if '+' in x:
        val=x.replace('+','')
    if ',' in x:
        val=x.replace(',','')
    else:
        val=0
    return val    
```

```python
df_dropna['Installs']=df_dropna['Installs'].apply(abc)
```
